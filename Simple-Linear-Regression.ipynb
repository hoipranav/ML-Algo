```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class MyLinearRegression:
    def __init__(self):
        self.m = 0
        self.c = 0
        self.y_avg = 0
        self.x_avg = 0

    def calculate_m(self, X, y):
        numerator = 0
        denominator = 0
        for i in range(X.shape[0]):
            numerator += (X[i] - self.x_avg) * (y[i] - self.y_avg)
            denominator += (X[i] - self.x_avg) ** 2
        return numerator / denominator

    def calculate_c(self):
        return (self.y_avg - (self.m * self.x_avg))

    def fit(self, X, y):
        self.x_avg = np.mean(X)
        self.y_avg = np.mean(y)
        self.m = self.calculate_m(X, y)
        self.c = self.calculate_c()
        return self

    def predict(self, x):
        return ((self.m * x) + self.c)
```

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=100, n_features=1, noise=80, random_state=42)
```

```python
plt.scatter(X, y)
```

```python
lr = MyLinearRegression()
```

```python
lr.fit(X, y)
```

```python
y_pred = lr.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color="red")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
olr = LinearRegression()
```

```python
olr.fit(X, y)
```

```python
olr_pred = olr.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, olr_pred, color="red")
```
