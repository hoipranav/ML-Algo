```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=100, n_features=1, noise=80, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
lr_pred = lr.predict(X)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
plt.scatter(X, y)
plt.plot(X, lr_pred, color="red")
```

```python
class MyGradientDescent:
    def __init__(self, learning_rate, epochs):
        self.m = 1
        self.b = 0
        self.lr = learning_rate
        self.epochs = epochs
        self.y = None

    def fit(self, X, y):
        self.y = y
        return self

    def predict(self, X):
        for i in range(self.epochs):
            loss_b_slope = -2 * np.sum(self.y - (self.m * X.ravel()) - self.b)
            loss_m_slope = -2 * np.sum((self.y - (self.m * X.ravel()) - self.b) * X.ravel())
            stepsize_for_b = self.lr * loss_b_slope
            stepsize_for_m = self.lr * loss_m_slope
            self.b = self.b - stepsize_for_b
            self.m = self.m - stepsize_for_m
            y_pred = (self.m * X.ravel()) + self.b
        return y_pred
```

```python
gd = MyGradientDescent(0.001, 100)
```

```python
gd.fit(X, y)
```

```python
y_pred = gd.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, lr_pred, color="red")
plt.plot(X, y_pred, color="blue")
```
